Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose

Simple Linear Regression (SLR) is a statistical and machine learning technique used to study and predict the relationship between one independent variable (X) and one dependent variable (Y). It assumes that the relationship between the two variables is linear, meaning it can be represented by a straight line.

The mathematical equation of Simple Linear Regression is:

Y=a+bX

Where:

Y = Dependent variable (the value to be predicted)
X = Independent variable (the predictor)
a = Intercept (the value of Y when X = 0)
b = Slope or coefficient (the amount by which Y changes for a one-unit change in X)
Purpose of Simple Linear Regression

The main purposes of SLR are:

To understand the relationship between two variables.
To predict future values of the dependent variable based on the independent variable.
To measure the impact of the independent variable on the dependent variable.
To identify trends and patterns in data

Question 2: What are the key assumptions of Simple Linear Regression?

The key assumptions of Simple Linear Regression (SLR) are:

Linearity
There should be a linear relationship between the independent variable (X) and the dependent variable (Y). This means that changes in X produce proportional changes in Y.
Independence of Errors
The observations and residuals (errors) should be independent of each other. The error associated with one observation should not influence another.
Homoscedasticity
The variance of the residuals should remain constant across all values of the independent variable. In other words, the spread of errors should be approximately the same throughout the dataset.
Normality of Residuals
The residuals (prediction errors) should be normally distributed. This assumption is important for statistical inference and hypothesis testing.
No Significant Outliers
The dataset should not contain extreme outliers that can disproportionately affect the regression line and lead to inaccurate results.
Measurement Without Error in the Independent Variable
The independent variable (X) is assumed to be measured accurately, with minimal or no measurement error.

Question 3: Write the mathematical equation for a simple linear regression model and explain each term.

The mathematical equation for a Simple Linear Regression (SLR) model is:

Y=β0+β1X+ε

Where:

Y = Dependent variable (response or output variable) that we want to predict.
X = Independent variable (predictor or input variable) used to make the prediction.
β0 = Intercept or constant term, which represents the value of Y when X=0.

β1 = Slope coefficient, which indicates the amount of change in Y for a one-unit increase in X.
ε = Error term (residual), which represents the difference between the actual value and the predicted value due to factors not included in the model.

The predicted equation is often written as:

Y^=b0+b1X

where:

Y^ = Predicted value of the dependent variable

b0​ = Estimated intercept

b1 = Estimated slope

Question 8: What is the purpose of the R-squared metric in regression analysis?

R-squared (R²), also known as the coefficient of determination, is a metric used in regression analysis to measure how well the regression model explains the variability in the dependent (target) variable.

The purpose of the R² metric is to indicate the proportion of the variance in the actual data that is explained by the independent variables in the model. Its value ranges from 0 to 1:

R² = 1: The model perfectly explains all the variability in the data.
R² = 0: The model does not explain any of the variability in the data.
Higher R² values indicate a better fit of the model to the data.

Formula:

R = 1- (yi - y^ i)

```python
# Import required libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable
y = np.array([2, 4, 5, 4, 5])                 # Dependent variable

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print the slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

Question 10: How do you interpret the coefficients in a simple linear regression model?

In a Simple Linear Regression model, the equation is expressed as:

**Y = β₀ + β₁X + ε**

where:

* **Y** is the dependent variable (response variable),
* **X** is the independent variable (predictor variable),
* **β₀** is the intercept,
* **β₁** is the slope coefficient, and
* **ε** represents the error term.

The **intercept (β₀)** represents the predicted value of the dependent variable when the independent variable is equal to zero. It indicates the starting point of the regression line.

The **slope coefficient (β₁)** represents the expected change in the dependent variable for every one-unit increase in the independent variable. If the coefficient is positive, the dependent variable increases as the independent variable increases. If the coefficient is negative, the dependent variable decreases as the independent variable increases.

For example, if the regression equation is:

**Salary = 20,000 + 5,000 × Experience**

then:

* The intercept (20,000) indicates that a person with zero years of experience is predicted to earn ₹20,000.
* The slope (5,000) indicates that for each additional year of experience, the salary is expected to increase by ₹5,000.
